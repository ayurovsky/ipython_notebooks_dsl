```python
import numpy as np
import matplotlib.pyplot as plt


N=3

diwan_percent_depletion = (176/213, 42/176, 34/176)
diwan_error = (0.026, 0.027, 0.025)

ind = np.arange(N)  # the x locations for the groups
width = 0.35       # the width of the bars

fig, ax = plt.subplots(figsize=(9, 6))
rects1 = ax.bar(ind, diwan_percent_depletion, width, color='#666666', yerr=diwan_error, capsize=3, ecolor='#A9A9A9')

ruhul_percent_depletion = (84/153, 35/153, 9/153)
ruhul_error = (0.04, 0.034, 0.019)
rects2 = ax.bar(ind + width, ruhul_percent_depletion, width, color='#999999', yerr=ruhul_error, capsize=3, ecolor='#A9A9A9')

# add some text for labels, title and axes ticks
ax.set_ylabel('Percent Species Depleted', fontsize=18)
ax.set_title('Percent Species Depleted by Method and Dataset', fontsize=20)
ax.set_xticks(ind + width / 2)
ax.set_xticklabels(('Diwan high affinity\ncodon shuffle', 'Complementary\ncodon shuffle', 'Combined P-val'), fontsize=15)

ax.legend((rects1[0], rects2[0]), ('Diwan Dataset Perfect aSD', 'Control Dataset No aSD'), fontsize=15, bbox_to_anchor=(1, 0.95), frameon=False)
plt.tick_params(axis='both', which='major', labelsize=15)

# Hide the right and top spines
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)

plt.show()
```
